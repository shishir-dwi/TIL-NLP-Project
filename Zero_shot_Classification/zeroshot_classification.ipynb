```python
output = []

filter_key1 = ['labels']
filter_key2 = ['scores']

for index, row in dff.iterrows():
  d = {}
  seq = row['data']
  result = classifier(seq,sample, multi_label=True)
 # print(result)
  temp_label = list(map(result.get, filter_key1))
  temp_score = list(map(result.get, filter_key2))
 # print(temp)
  d['headline_text'] = row['data']
  d['label'] = temp_label[0][0]
  d['score'] = temp_score[0][0]
  output.append(d) 
```

```python
output = []
count = 0  # Initialize counter

filter_key1 = ['labels']
filter_key2 = ['scores']

for index, row in dff.iloc[452:].iterrows():
    d = {}
    seq = row['data']
    result = classifier(seq, sample, multi_label=True)

    temp_label = list(map(result.get, filter_key1))
    temp_score = list(map(result.get, filter_key2))

    d['headline_text'] = row['data']
    d['label'] = temp_label[0][0]
    d['score'] = temp_score[0][0]
    output.append(d)

    # Increment the counter
    count += 1

    # Print the count
    print(f"Processed {count} rows")

# After the loop, you can print the final count
print(f"Total rows processed: {count}")
```
