```python
import pandas as pd
df=pd.read_csv('Embedding - Fasttext.csv')
df.head()
```

```python
columns_to_keep = ['embedding', 'target']
df = df[columns_to_keep]
```

```python
df.head()
```

```python
type(df.iloc[0].embedding)
```

```python
categories=df.target.unique()
categories
```

```python
sample_df = pd.DataFrame(columns=['embedding', 'target'])
for category in categories:
    category_samples = df[df['target'] == category].sample(5000)
    sample_df = pd.concat([sample_df, category_samples])
sample_df.reset_index(drop=True, inplace=True)
sample_df
```

```python
import numpy as np
def text_to_list(embedding_text):
                 embedding_text=embedding_text.strip('[]')
                 return np.array([float(val) for val in embedding_text.split()])
sample_df['embedding']=sample_df['embedding'].apply(text_to_list)
```

```python
type(sample_df.iloc[0].embedding)
```

```python
from sklearn.model_selection import train_test_split
x = np.array(sample_df['embedding'].tolist())
y = sample_df['target'].tolist()
```

```python
x.shape
```

```python
x=x.reshape(130000,1,300)
x.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,train_size=0.8, stratify=y, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Flatten
import time
from memory_profiler import memory_usage


model = Sequential()
model.add(LSTM(512, return_sequences=True, input_shape=(1, 300)))
model.add(Dropout(0.3))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(128, return_sequences=True))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(26, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()


training_time=0

def train_model():
    global training_time
    start_time = time.time()
    model.fit(x_train, y_train, validation_split=0.1, epochs=100, batch_size=64)
    end_time = time.time()
    training_time = end_time - start_time
mem_usage = memory_usage(train_model)
max_mem_usage = max(mem_usage)

print("MEMORY USAGE USING LSTM:- ",max_mem_usage)
print("TRAINING TIME(LSTM):- ",training_time)
```

```python
model.save('lstm_model.h5')
```

```python
from sklearn.metrics import accuracy_score, classification_report

start_time = time.time()
y_pred=model.predict(x_test)
predicted_classes = y_pred.argmax(axis=1)
end_time = time.time()
inference_time = end_time - start_time
print("INFERENCE TIME USING LSTM:- ",inference_time)
accuracy=accuracy_score(y_test,predicted_classes)
print(f'ACCURACY:- {accuracy:.2f}')
classification_rep = classification_report(y_test, predicted_classes)
print('\nCLASSIFICATION REPORT:- ')
print(classification_rep)
```
